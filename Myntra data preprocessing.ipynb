```python
data.drop('Total Reviews', axis = 1)
```

```python
# Split the 'ratings' column into 'rating' and 'total_reviews'
data[['rating', 'total_reviews']] = data['ratings'].str.split('|', expand=True)
```

```python
# List of columns to drop
columns_to_drop = ['Total Reviews', 'ratings']

# Drop the specified columns
df = data.drop(columns=columns_to_drop, axis=1)
```

```python
df
```

```python
df['total_reviews'].unique()
```

```python
data1 = df

data1['total_reviews'] = df['total_reviews'].replace({r'k$': 'e3'}, regex=True).apply(
    lambda x: eval(x) if pd.notna(x) else x
)

data1
```
